```python
import spacy
#from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cosine
```

```python
# Загружаем модель русского языка
nlp = spacy.load("ru_core_news_md")
```

```python
# Мой текст
text = '''Там за лесом, большая поляна. Я ложусь на нее, и смотрю в небо, наслаждаясь пением птиц и шумом листьев. Так можно весь день пролежать, отпуская мысли на свободу, играя с ветром, с его могуществом. А за поляной, в двух шагах, течет маленькая речка. Вода в ней чистая и очень свежая. Наберешь ее в ладони, и умоешь лицо. И как заново родился. Течение несет ее куда-то дальше, свозь камни и ветви, упавшие с деревьев. Закроешь тогда глаза, вдохнешь эти запахи, и улыбнешься. Ведь когда еще будет возможность, столь глубоко сблизиться с самой природой. Я опускаю ладони в реку и вода тут же охватывает мои пальцы. Вся ее энергия  просачивается сквозь мою кожу и пропитывает меня насквозь.'''
```

```python
# Обработаем текст
doc_text = nlp(text)
print(doc_text)
```

```python
# Вывод списка стоп-слов по умолчанию в Spacy
stop_words = nlp.Defaults.stop_words
stop_words_list = sorted(list(stop_words))
print("Список стоп-слов:")
for word in stop_words_list:
    print(word)
```

```python
# Новый текст после прохождения через список стоп-слов
filtered_words = [token.text for token in doc_text if not token.is_stop]
filtered_text = ' '.join(filtered_words)
# Вывод нового текста
print(filtered_text)
```

```python
doc_filtered_text = nlp(filtered_text)
```

```python
#Отобразим векторы для исходного текста и обработанного текста
vec_text = doc_text.vector
vec_filtered_text = doc_filtered_text.vector
```

```python
print(vec_text)
```

```python
# Вектор после прохождения стоп-слов
print(vec_filtered_text)
```

```python
# рассчитаем косинусную меру близости между начальным вектором и вектором после прохождения стоп-слов :
from scipy.spatial.distance import cosine

similarity = 1 - cosine(vec_text, vec_filtered_text)
print(f"\n Косинусная мера близости= {similarity:.4f}")
```

```python
# Сравнение близости векторов. Для проверки
print(doc_text.similarity(doc_filtered_text))
```

```python
# Разбивка текста на отдельные предложения
sentences = list(doc_text.sents)
for sentence in sentences:
    print (sentence.text)
```

```python
# Каждое предложение пропущу через стоп-слова
def remove_stop_words(sentence):
    return ' '.join([word for word in sentence.split() if word.lower() not in stop_words])
for sentence in sentences:
    cleaned_sentence = remove_stop_words(sentence.text)
    print(cleaned_sentence)
```

```python
doc_cleaned_sentence = nlp(cleaned_sentence)
print(doc_cleaned_sentence)
```

```python
# Получим вектора каждых предложений
vec_cleaned_sentence = doc_cleaned_sentence.vector
#print(doc_cleaned_sentence)
print(vec_cleaned_sentence)
```
