```python
# Импортируем библиотеку spacy
import spacy
```

```python
# загружаем предобученную модель ru_core_news_md для русского языка из библиотеки Spacy и выводим список компонентов, которые включены в текущую модель
nlp = spacy.load("ru_core_news_md")
print(nlp.pipe_names)
```

```python
# функция принимает путь к файлу (path) и открывает файл в режиме чтения ('r'), используя кодировку UTF-8
def read_file(path):
    with open(path, 'r', encoding='utf-8') as file:
        return file.read()
        
file_path_50_00_2020 = read_file('/home/datadisk/jupyter-usmanov_il/UsmanovIL/UsmanovIL/Векторизация_текста3/50_00_2020.txt')
print(file_path_50_00_2020)
```

```python
# выполняется обработка текста
doc_50_00_2020 = nlp(file_path_50_00_2020)
print(doc_50_00_2020)
```

```python
# переменную stop_words_list, в которой хранится отсортированный список стоп-слов. Для этого мы используем функцию list(), 
# чтобы превратить множество стоп-слов в список, и функцию sorted(), чтобы отсортировать элементы списка в алфавитном порядке
stop_words = nlp.Defaults.stop_words
stop_words_list = sorted(list(stop_words))
print(stop_words_list)
```

```python
# перебираю все элементы списка stop_words_list и вывожу каждый элемент на отдельную строку
for word in stop_words_list:
    print(word)
```

```python
# удаляем из документа все стоп-слова и формируем новые строки, состоящие только из значимых слов
filtered_words = [token.text for token in doc_50_00_2020 if not token.is_stop]
filtered_text = ' '.join(filtered_words)
print(filtered_text)
```

```python
# разобьем текст на предложения и сохраним их в виде списка.
sentences = list(doc_50_00_2020.sents)
print(sentences)
```

```python
# тоже разбиение текста на предложения
sentences = [sent.text for sent in doc_50_00_2020.sents]
print(sentences)
```

```python
# Тоже самое. Для сверки. Теперь у нас есть список предложений sentences, где каждое предложение - это строка.
sentences = []
for sent in doc_50_00_2020.sents:
    sentences.append(sent.text.strip())
print(sentences)
```

```python
# Преобразуем каждое предложение в вектор
sentence_vectors = []
for sentence in sentences:
    doc_sentence = nlp(sentence)
    vector = doc_sentence.vector
    sentence_vectors.append(vector)
print(sentence_vectors)
```

```python
# Тоже самое. Для сверки делал. Преобразуем каждое предложение в вектор
vec_sentences = []
for sentence in sentences:
    doc_sentence = nlp(sentence)
    vec_sentences.append(doc_sentence.vector)
print(vec_sentences)
```

```python
# Создадим словарь, который будет содержать информацию о каждом предложении: его идентификатор, исходный текст и соответствующий вектор
document_structure = {}
for i, (sentence, vector) in enumerate(zip(sentences, sentence_vectors)):
    document_structure[i] = {
        'id': i,
        'text': sentence,
        'vector': vector
    }
print(document_structure)
```
