```python
# Функция для удаления стоп-слов
def remove_stopwords(text):
    doc = nlp(text)
    filtered_tokens = [token.text for token in doc if not token.is_stop]
    return ' '.join(filtered_tokens)

# Чтение и очистка всех файлов
documents = []
for path in file_paths:
    text = read_file(path)
    cleaned_text = remove_stopwords(text)
    documents.append(cleaned_text)

print(cleaned_text)
```

```python
# Функция для разбиения текста на предложения и создания векторов
def vectorize_sentences(document):
    sentences = list(nlp(document).sents)
    vectors = [sent.vector for sent in sentences]
    return sentences, vectors

# Создание списка предложений и соответствующих им векторов для каждого документа
document_sents = []
document_vectors = []
for document in documents:
    sents, vectors = vectorize_sentences(document)
    document_sents.append(sents)
    document_vectors.append(vectors)
```

```python
# Структура для хранения информации о каждом предложении
data_structure = []

for i, (sents, vectors) in enumerate(zip(document_sents, document_vectors)):
    for j, (sent, vector) in enumerate(zip(sents, vectors)):
        data_structure.append({
            'document_id': i + 1,
            'sentence_id': j + 1,
            'vector': vector
        })
```

```python
# Выбираем одно из предложений в качестве поискового запроса
search_query = document_sents[0][25].text
print(search_query)

# Векторизуем поисковый запрос
query_vector = nlp(search_query).vector
print(query_vector)
```

```python
# Находим ближайшее предложение по косинусному расстоянию
closest_sentence = None
min_distance = float('inf')

for item in data_structure:
    # Проверяем, чтобы длина вектора была больше нуля
    if np.linalg.norm(item['vector']) > 0:
        distance = cosine(query_vector, item['vector'])
        if distance < min_distance:
            closest_sentence = item
            min_distance = distance
    else:
        print(f"Вектор предложения {item['document_id']}:{item['sentence_id']} имеет нулевую длину.")

if closest_sentence is not None:
    print(f'Ближайшее предложение: {closest_sentence["document_id"]}:{closest_sentence["sentence_id"]}')
else:
    print("Не найдено ни одного подходящего предложения.")
```

```python
# Меняем текст запроса, сохраняя смысл
modified_search_query = ' В настоящем стандарте под термином "информационная система" понимается автоматизированная информационная система. '

# Векторизуем изменённый запрос
modified_query_vector = nlp(modified_search_query).vector

# Находим ближайшее предложение по косинусному расстоянию для изменённого запроса
closest_modified_sentence = None
min_modified_distance = float('inf')

for item in data_structure:
    distance = cosine(modified_query_vector, item['vector'])
    if distance < min_modified_distance:
        closest_modified_sentence = item
        min_modified_distance = distance

print(f'Ближайшее предложение для измененного запроса: {closest_modified_sentence["document_id"]}:{closest_modified_sentence["sentence_id"]}')
```
