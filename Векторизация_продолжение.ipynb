```python
file_path_50_00_2020 = '''Приложение
к приказу ПАО «Сургутнефтегаз»
от «___»________20__ г. №_____











СТАНДАРТ ОРГАНИЗАЦИИ
__________________________________________________________________
ЕДИНАЯ ИНФОРМАЦИОННАЯ СИСТЕМА 
ПАО «СУРГУТНЕФТЕГАЗ»
Основные положения
СТО 50.0–2020
(Взамен СТО 50.0–2011)
Предисловие
    1 РАЗРАБОТАН управлением информационных технологий ПАО «Сургутнефтегаз» 
    2 УТВЕРЖДЕН И ВВЕДЕН В ДЕЙСТВИЕ приказом ПАО «Сургутнефтегаз» от «____»_________ 20__ г. №________
    3 ВЗАМЕН СТО 50.0–2011



































© ПАО «Сургутнефтегаз», 2020
Настоящий стандарт не может быть полностью или частично воспроизведен, тиражирован и распространен без письменного согласия ПАО «Сургутнефтегаз»
Содержание
1	Область применения	12
2	Нормативные ссылки	12
3	Термины и определения	2
4	Основные положения	62
5	Жизненный цикл ИТ-решения	62
6	Процессы управления жизненным циклом	102
7	Система стандартов и организационно-технологических документов «Единой информационной системы ПАО «Сургутнефтегаз»	112
8	Группы ролей и участники	142
Приложение А (справочное)  Единая информационная система Общества	162
Приложение Б (справочное)  Процессы управления жизненным циклом 
 ИТ-решения	172
Лист регистрации изменений СТО 50.0–2020	202



СТАНДАРТ ОРГАНИЗАЦИИ
ЕДИНАЯ ИНФОРМАЦИОННАЯ СИСТЕМА ПАО «СУРГУТНЕФТЕГАЗ»
Основные положения
СТО 50.0–2020

Взамен СТО 50.0–2011
Дата введения_______________________
    1 Область применения
        1.1 Настоящий стандарт устанавливает основные положения по управлению ИТ-решениями в единой информационной системе (ЕИС) публичного акционерного общества «Сургутнефтегаз» (далее по тексту – Общество).
        1.2 Стандарт обязателен для всех структурных подразделений Общества, а также управлений, отделов и служб аппарата управления Общества.
        1.3 Специалисты Общества, принимавшие участие в подготовке, пересмотре, корректировке, отмене, в методическом обеспечении, согласовании и нормоконтроле стандартов в пределах своих должностных обязанностей и направлений деятельности несут ответственность, установленную действующим законодательством Российской Федерации, за корректность изложения стандартов Общества и их соответствие требованиям действующего законодательства Российской Федерации и локальным актам Общества. 
    2 Нормативные ссылки
В настоящем стандарте использованы нормативные ссылки на следующие стандарты:
ГОСТ 34.603–92 Информационная технология. Виды испытаний автоматизированных систем
ГОСТ 19781–90 Обеспечение систем обработки информации программное. Термины и определения
ГОСТ Р 57193–2016 Системная и программная инженерия. Процессы жизненного цикла систем 
ГОСТ Р ИСО/МЭК 20000-1–2013 Информационная технология. Управление услугами. Часть 1. Требования к системе управления услугами
ГОСТ ISO 9000–2011 Системы менеджмента качества. Основные положения и словарь
СТО 4–2018 Нормативно-технические документы ОАО «Сургут-нефтегаз». Виды, порядок разработки, оформления, регистрации, утверждения, учета, обновления и отмены
СТО 10.2–2018 Система обеспечения информационной безопасности ПАО «Сургутнефтегаз». Документы информационной безопасности ПАО «Сургутнефтегаз»
Примечание – При пользовании настоящим стандартом целесообразно проверить действие ссылочных стандартов и классификаторов в информационной системе общего пользования – на официальном сайте Федерального агентства по техническому регулированию и метрологии в сети Интернет, в информационно-поисковой системе «Стандарт+», в «Базе данных «СТО ПАО «Сургутнефтегаз» в разделе «Техническое управление» корпоративного портала Общества. Если ссылочный стандарт заменен (изменен), то при пользовании настоящим стандартом следует руководствоваться заменяющим (измененным) стандартом. Если ссылочный стандарт отменен без замены, то положение, в котором дана ссылка на него, применяется в части, не затрагивающей эту ссылку.
    3 Термины и определения
        3.1 В настоящем стандарте применены следующие термины с соответствующими определениями:
            3.1.1 бизнес-архитектура: Целостная организация бизнес-процессов, описывающих деятельность Общества.
            3.1.2 бизнес-цели; БЦ: Цели, на достижение которых направлена производственно-хозяйственная деятельность Общества.
            3.1.3 бизнес-процесс (процесс): Устойчивая совокупность упорядоченных, взаимосвязанных действий, направленных на достижение значимого результата при осуществлении производственно-хозяйственной деятельности Общества.
            3.1.4 Владелец бизнес-процесса: Работник или группа работников, отвечающих за исполнение и результаты процесса и имеющих полномочия изменять его с целью усовершенствования, в том числе посредством предоставленной роли.
            3.1.5 данные: Представление информации в формализованном виде, пригодном для передачи, интерпретации или обработки. 
            3.1.6 договор: Соглашение между двумя сторонами об установлении, изменении или прекращении прав и обязанностей, оформленное в соответствии с действующим законодательством Российской Федерации.
Примечание – В настоящем стандарте подразумевается договор по предоставлению услуги на поставку, разработку, эксплуатацию или сопровождение ИТ-решения.
            3.1.7 Заказчик: Структурное подразделение или управление, отдел, служба аппарата управления Общества, которая приобретает или получает готовое для эксплуатации ИТ-решение от Исполнителя или Поставщика.
            3.1.8 информационная безопасность: Все аспекты, связанные с определением, достижением и поддержанием конфиденциальности, целостности, доступности, отказоустойчивости, подотчетности, аутентичности и достоверности информации или средств ее обработки.
            3.1.9 информационная система; ИС: Программно-аппаратная система, предназначенная для автоматизации целенаправленной деятельности конечных пользователей, обеспечивающая в соответствии с заложенной в нее логикой обработки возможность получения, модификации и хранения информации.
Примечание – Термин «информационная система» в настоящем стандарте рассматривается как автоматизированная информационная система.
            3.1.10 информационные технологии; ИТ: Технологии хранения, преобразования, защиты, обработки, передачи и получения информации с использованием вычислительной техники и программного обеспечения. 
            3.1.11 инцидент: Незапланированное событие, которое привело или может привести к прерыванию предоставления услуги или к снижению ее качества, даже если оно еще не повлияло на услугу для Заказчика (ГОСТ Р ИСО/МЭК 20000-1, пункт 3.10). 
            3.1.12 Исполнитель: Структурное подразделение Общества либо его организационная единица, управление, отдел или служба аппарата управления Общества, предоставляющие услуги по проектированию, реализации, внедрению, эксплуатации и сопровождению ИТ-решения.
            3.1.13 ИТ-архитектура: Логика организации данных, приложений, инфраструктуры, определенная в наборе методик, взаимосвязей и технических операций для достижения желаемых показателей, технической стандартизации и интеграции.
            3.1.14 ИТ-инфраструктура: Совокупность технического и программного обеспечения, обеспечивающего работу приложений, а также организационных, правовых и нормативных механизмов, обеспечивающих их функционирование.
Примечания
1 Техническое обеспечение состоит из серверов, персональных компьютеров, систем хранения данных, сети и коммуникационных приложений. 
2 Программное обеспечение характеризуется операционными системами, инструментальными средами разработки и средствами обеспечения информационной безопасности.
            3.1.15 ИТ-решение: Автоматизированный в одной или нескольких информационных системах бизнес-процесс (функции процесса), построенный на освоенных информационных технологиях и удовлетворяющий требованиям Заказчика.
            3.1.16 ИТ-услуга: Выполнение работ, заданий в рамках деятельности по освоению и использованию информационных технологий, направленных на удовлетворение Исполнителем (Поставщиком) потребностей Заказчика.
            3.1.17 ИТ-служба: Штатная организационная единица Общества или группа работников, формируемая на длительной (постоянной) основе, в положениях о которой присутствуют функции создания, изменения, сопровождения ИТ-решений (ПО, программного средства, ИС и пользователей) или компонентов ИТ-инфраструктуры в рамках развития ЕИС Общества.
            3.1.18 ИТ-процесс: Бизнес-процесс либо технологический процесс, связанный с деятельностью в области использования и управления информационными технологиями.
            3.1.19 компонент ИТ-решения: Представитель совокупности элементов, образующих ИТ-решение.
Примечание – Компоненты могут разниться по сложности, размерам и типам от ИС до отдельного модуля или аппаратного компонента.
            3.1.20 модель жизненного цикла: Структурная основа процессов и действий, относящихся к жизненному циклу, которая также служит в качестве общего эталона для установления связей и понимания 
(ГОСТ Р 57193, пункт 4.1.20).
            3.1.21 мониторинг: Процесс систематического или непрерывного сбора информации о параметрах функционирования объектов мониторинга для определения тенденций изменения параметров.
            3.1.22 организационно-технологические документы; ОТД: Документы, в которых содержатся решения по организации деятельности и технологии производства работ в области ИТ для развития ЕИС Общества. 
            3.1.23 показатель: Индикаторы, выраженные в абсолютных или относительных величинах, характеризующих результат измерений или процедуры интерпретации оценки.
            3.1.24 Пользователь: Работник, который использует действующее 
ИТ-решение для выполнения конкретной трудовой функции.
Примечание – Пользователями могут выступать работники со стороны Заказчика и Исполнителя.
            3.1.25 Поставщик: Организация, не входящая в структуру Общества, предоставляющая готовое ИТ-решение или ИТ-услугу в рамках заключенных договоров. 
            3.1.26 проблема: Корневая причина одного или нескольких инцидентов (ГОСТ Р ИСО/МЭК 20000-1, пункт 3.19).
Примечание – Как правило, корневая причина неизвестна на момент создания записи о проблеме, и дальнейшее исследование проводится в рамках процесса управления проблемами.
            3.1.27 проект: Уникальный процесс, состоящий из совокупности скоординированных и управляемых видов деятельности с начальной и конечной датами, предпринятый для достижения цели, соответствующей конкретным требованиям, включающий ограничения по срокам, стоимости и ресурсам (ГОСТ ISO 9000, пункт 3.4.3).
            3.1.28 приложение (прикладное программное обеспечение): Компьютерная программа (или группа программ), предназначенная для выполнения определенных пользовательских задач и рассчитанная на непосредственное взаимодействие с Пользователем. 
            3.1.29 проектно-техническая документация; ПТД: Комплекс взаимосвязанных документов, полностью определяющих спецификацию требований на уровне бизнес-задач, проектные и организационные решения по их реализации, готовности ИТ-решения к эксплуатации (функционированию).
            3.1.30 программное обеспечение; ПО: Совокупность программ системы обработки информации и программных документов, необходимых для эксплуатации этих программ (ГОСТ 19781).
            3.1.31 программное средство: Объект, состоящий из программ, процедур, правил, а также, если предусмотрено, сопутствующих им документации и данных, относящихся к функционированию системы обработки информации (ГОСТ 19781).
            3.1.32 роль: Набор обязанностей, действий и полномочий, предоставленных работнику или группе работников.
            3.1.33 система управления: Структура, состоящая из взаимосвязанных политик, процессов и функций, которая гарантирует достижение определенных целей.
            3.1.34 требование: Потребность или ожидание, которое установлено, обычно предполагается или является обязательным (ГОСТ ISO 9000, 
пункт 3.1.2).
            3.1.35 цель: Желаемое будущее состояние, которое может быть достигнуто при выполнении выделенного состава функций с принятием во внимание критических факторов успеха. 
            3.1.36 опытная эксплуатация: Испытание ИТ-решения, проводимое в целях определения фактических значений количественных и качественных характеристик ИТ-решения и готовности персонала к работе в условиях функционирования ИТ-решения, корректировки (при необходимости) документации (с учетом ГОСТ 34.603).
            3.1.37 опытно-промышленная эксплуатация: Опытная эксплуатация, в процессе которой результаты работы ИТ-решения используются в производственно-хозяйственной деятельности Общества, время реагирования на инциденты приближено ко времени, регламентированному для эксплуатируемых ИТ-решений.
            3.1.38 промышленная эксплуатация: Использование ИТ-решения в производственно-хозяйственной деятельности Общества с принятием бизнес-решений на основе результатов его работы.
            3.1.39 прототип: Предварительный образец или пример системы, который служит в качестве модели на более поздних этапах жизненного цикла системы или для окончательной, завершенной версии системы.

    4 Основные положения
        4.1 ЕИС Общества представляет собой совокупность систем управления во взаимосвязи и взаимодействии, объединенных регулируемыми взаимоотношениями для осуществления целенаправленной деятельности Общества (приложение А).
        4.2 Бизнес-архитектура ЕИС Общества определяет общую структуру и функции систем управления, описывает совокупность взаимосвязанных бизнес-процессов деятельности Общества, включая организационные и функциональные структуры, роли и ответственности, стандарты и руководства.
        4.3 ИТ-архитектура ЕИС Общества описывает логику организации и интеграции ИС, обеспечивает соответствие бизнес-задач по обработке информации возможностям ИТ. Решения по автоматизации бизнес-процессов на базе действующих ИС в дальнейшем понимаются как ИТ-решения.
        4.4 Планируемые изменения бизнес-архитектуры Общества и ИТ обеспечивают целенаправленные процессы перевода ИТ-архитектуры из текущего состояния в желаемое будущее.
        4.5 Архитектура ИС и ИТ-решений соответствует и существует в рамках ИТ-архитектуры – логической взаимосвязи компонентов ИТ-архитектуры – данных, приложений, элементов ИТ-инфраструктуры.
        4.6 Архитектура информации (данных) определяет, какие данные необходимы для поддержания бизнес-процессов и их использования в ИС.
        4.7 Архитектура приложений определяет, какие приложения используются и должны использоваться для управления данными и поддержки бизнес-процессов.
        4.8 Технологическая архитектура (архитектура ИТ-инфраструктуры) определяет, какое аппаратное и системное ПО, какие сети и коммуникации необходимы для создания среды работы приложений, которые, в свою очередь, управляют данными и обеспечивают исполнение бизнес-процесса.
        4.9 ИТ-решениям любого типа, размера и сложности свойственны основные этапы эволюции от установления требований к ним до прекращения их использования. В основе процесса эффективного управления 
ИТ-решениями лежит понятие жизненного цикла. 
    5 Жизненный цикл ИТ-решения
        5.1 Основные процессы жизненного цикла 
            5.1.1 Жизненный цикл ИТ-решения основывается на модели жизненного цикла и включает в себя этапы от установления требований Заказчика к автоматизации и реализации ИТ-решения до вывода его из эксплуатации, то есть до прекращения использования (приложение Б). 
            5.1.2 Основными этапами жизненного цикла ИТ-решения являются:
- планирование;
- поставка;
- проектирование;
- реализация;
- заключительная подготовка и внедрение;
- эксплуатация и сопровождение;
- вывод из эксплуатации.
        5.2 Планирование
            5.2.1 На этапе планирования ИТ-решения (проработки заказа на 
ИТ-решение в области автоматизации процессов или оптимизации работ) определяются цели, проблемы и потребности Заказчика. 
            5.2.2 Исполнителем уточняются требования Заказчика, оцениваются возможности собственной реализации, покупки готового ИТ-решения, учитываются риски, стоимость, необходимые ресурсы, планируется выполнение проектных работ.
            5.2.3 Заказчик или Исполнитель (от имени Заказчика) должен учесть в требованиях Заказчика особенности лицензирования (права собственности, способы использования, гарантии, связанные с использованием готовых программных продуктов), первоначальное и непрерывное обучение, техническую поддержку персонала Заказчика.
            5.2.4 Заказчиком (при необходимости, совместно с Исполнителем) при приобретении готового ИТ-решения (или части решения) выбирается Поставщик, подготавливается, а в последующем контролируется выполнение условий договора с Поставщиком на поставку решения или услуг. 
            5.2.5 Заказчик или Исполнитель (от имени Заказчика) должен контролировать наличие исходных требований Заказчика в заключаемых договорах с Поставщиком.
        5.3 Поставка
            5.3.1 Процесс поставки заключается в обеспечении Заказчиком или Исполнителем в лице Заказчика необходимыми условиями (например, уточнение постановки задачи, предоставление требований зафиксированных в регламентирующих документах к процессу и результату выполняемых работ, обеспечение подключения к системам разработки и тестирования при проведении работ на площадке Общества и т.п.) для выполнения поставки ИТ-решения или услуги Поставщиком с учетом условий заключенных договоров.
            5.3.2 Внедрение ИТ-решения, поставляемого в рамках договоров, осуществляется на этапе заключительной подготовки и внедрения (см. 5.6).
        5.4 Проектирование
            5.4.1 Проектирование ИТ-решения состоит из концептуального и технического проектирования, выполняемого Исполнителем.
            5.4.2 Технические решения в рамках проектирования документируются, согласовываются и утверждаются заинтересованными лицами, участвующими в управлении ИТ-решениями. Документально оформляются и решаются возникающие проблемы, устраняются несоответствия.
            5.4.3 Концептуальное проектирование.
                5.4.3.1 В соответствии с требованиями Заказчика Исполнителем прорабатываются функциональные требования к ИТ-решению, требования к данным, интерфейсам, безопасности, эргономике, эксплуатации. 
                5.4.3.2 Требования оцениваются на внутреннюю и внешнюю согласованность, тестируемость и выполнимость. 
                5.4.3.3 Исполнителем определяется архитектура ИТ-решения, которая описывает общую структуру ИТ-решения, объекты и конфигурации технических и программных средств, внешние и внутренние интерфейсы, оформляется ПТД.
            5.4.4 Техническое проектирование.
В процессе технического проектирования Исполнителем уточняются программные объекты на уровне программных модулей, прорабатывается разработка программных объектов и баз данных, определяются процедуры и критерии тестирования. 
Рекомендуется проведение демонстрации прототипа с учетом технической возможности используемых ИС.
        5.5 Реализация 
            5.5.1 В рамках реализации ИТ-решения Исполнителем выполняются работы по разработке программных объектов, сборке программных модулей, подготовке инфраструктуры для проведения тестирования.
            5.5.2 В процессе тестирования Исполнителем оцениваются результаты работ на соответствие требованиям, определенным Заказчиком к 
ИТ-решению на этапе планирования и уточненным Исполнителем на этапе проектирования.
        5.6 Заключительная подготовка и внедрение
            5.6.1 Для ввода ИТ-решения в эксплуатацию и его последующего сопровождения должна быть произведена оценка готовности инфраструктуры, в том числе и для готовых ИТ-решений, разработанных в рамках заключенных договоров.
            5.6.2 Исполнителем (Поставщиком) обеспечивается проведение Заказчиком оценки готовности ИТ-решения, фиксируется готовность 
ИТ-решения к промышленной эксплуатации. При анализе учитываются результаты тестирования на этапе реализации и путем проведения опытной или опытно-промышленной эксплуатации.
            5.6.3 В случае приобретения ИТ-решения Заказчик или Исполнитель
(от имени Заказчика) должен организационно и технически обеспечить соответствующую приемку услуг согласно договору.
            5.6.4 В процессе заключительной подготовки Исполнителем (Поставщиком) оформляется ПТД по поддержке инфраструктуры ИТ-решения и поддержке Пользователя, проводится обучение Пользователей. 
            5.6.5 Для внедрения (передача в промышленную эксплуатацию) и тиражирования ИТ-решения должны быть установлены процедуры получения и документирования сведений о возникших проблемах.
        5.7 Эксплуатация и сопровождение
            5.7.1 Поддержка Пользователя.
                5.7.1.1 На этапе эксплуатации Исполнитель обеспечивает помощь и консультации Пользователю.
                5.7.1.2 Запросы Пользователя и последующие ответные действия должны быть документально оформлены, контролируемы.
                5.7.1.3 При необходимости внесения изменения активизируется процесс управления изменениями, все принимаемые решения по изменению должны контролироваться вплоть до их выполнения.
            5.7.2 Поддержка инфраструктуры. 
На этапе эксплуатации Исполнителем:
- определяются работы по обслуживанию каждого элемента 
ИТ-решения в заданных условиях в интересах Пользователя и Заказчика в соответствии с ПТД;
- определяются процедуры мониторинга состояния, своевременного информирования и обработки сообщений о возникших инцидентах и проблемах в нормальном функционировании всех компонентов ИТ-решения.
            5.7.3 Управление изменениями.
                5.7.3.1 На этапе сопровождения Исполнитель контролирует изменения ИТ-решений и ПТД, вызванные возникшими проблемами или потребностями в доработке или настройке.
                5.7.3.2 Исполнителем анализируются полученные сообщения о возникших проблемах и заявки на внесение изменений, на основании полученных данных разрабатываются варианты реализации изменений, обеспечивается обратная связь с Пользователем и Заказчиком.
При необходимости, запросы переадресовываются Поставщику в рамках договоров.
                5.7.3.3 Исполнитель, в случае необходимости внесения изменений в действующие ИТ-решения, инициирует выполнение работ по проектированию (см. 5.4), реализации (см. 5.5), заключительной подготовке и внедрению (см. 5.6.) согласованных изменений. 
При внедрении изменений обеспечивается обучение Пользователя, обновление ПТД. 
        5.8 Вывод из эксплуатации
Для вывода из эксплуатации ИТ-решения Исполнитель должен разработать и довести до Пользователей план работ, содержание которого включает:
    • сроки прекращения полной или частичной поддержки ИТ-решения;
    • внесение изменений во взаимодействующие ИТ-решения;
    • обязательства по оставшимся вопросам поддержки ИТ-решения;
    • сроки перехода, при необходимости, к новому ИТ-решению;
    • требования по архивации ИТ-решения;
    • требования по доступу к архивным копиям данных;
    • требования к регистрации операций в бухгалтерском учете, связанных с выводом ИТ-решения.
    6 Процессы управления жизненным циклом
    5 
        6.1 Основные процессы управления жизненным циклом.
            6.1.1 Эффективное и качественное выполнение процессов жизненного цикла ИТ-решений обеспечивают: 
    • процесс управления требованиями;
    • процесс управления проектной и операционной деятельностью;
    • процесс управления ИТ-инфраструктурой;
    • процесс управления конфигурацией;
    • процесс управления качеством;
    • процесс управления обучающими материалами;
    • процесс обеспечения информационной безопасности;
    • процесс управления ИТ-услугами.
            6.1.2 Карта процессов управления ИТ-решениями в рамках жизненного цикла представлена в приложении Б.
        6.2 Процесс управления требованиями предназначен для определения требований Заказчика и преобразования этих требований в эффективные решения в области ИТ.
        6.3 Процесс управления проектной и операционной деятельностью.
            6.3.1 Проектирование, внедрение новых ИТ-решений и дополнительной функциональности к ним, удовлетворяющих потребностям Заказчика, выполняются в проектах. Проектная деятельность ограничена протяженностью во времени и направлена на достижение результата проекта. 
            6.3.2 Операционная деятельность является продолжающимися во времени и повторяющимися процессами на этапе эксплуатации и сопровождения. Работы выполняются выделенными группами, которые занимаются повседневной деятельностью, не ограниченной временными рамками.
        6.4 Процесс управления ИТ-инфраструктурой обеспечивает создание и сопровождение инфраструктуры, необходимой для любого этапа жизненного цикла. 
        6.5 Процесс управления конфигурацией предназначен для: 
- обозначения, определения и установления состояния (базовой линии) ПО в ИС; 
- управления изменениями и выпуском ПО; 
- обеспечения полноты и совместимости ПО; 
- управления хранением, обращением и поставкой ПО.
        6.6 Процесс управления качеством является процессом обеспечения соответствующих гарантий того, что ИТ-решение и процессы в жизненном цикле соответствуют установленным требованиям.
        6.7 Процесс управления обучающими материалами определяет работы по созданию материалов для Пользователей и Исполнителей и их использованию.
        6.8 Процесс обеспечения информационной безопасности обеспечивает идентификацию, контроль и защиту информации, а также любого оборудования, используемого для хранения, передачи и обработки информации. 
Данные процессы выполняются в соответствии с документами информационной безопасности, оформляемыми согласно СТО 10.2.
        6.9 Процесс управления ИТ-услугами решает задачи по классификации, планированию, внедрению, предоставлению, мониторингу и контролю, оценке качества ИТ-услуг, а также вопросы финансового управления и взаимоотношений с Заказчиком ИТ-услуг.
    7 Система стандартов и организационно-технологических 
документов «Единой информационной системы 
ПАО «Сургутнефтегаз» 
        7.1 Система стандартов и организационно-технологических документов «Единой информационной системы ПАО «Сургутнефтегаз» (далее по тексту – Система стандартов по ЕИС) направлена на проведение согласованной политики в области управления ИТ и является совокупностью основополагающих принципов и процедур по обеспечению развития и надежности эксплуатации ЕИС в рамках поддержания задач и целей Общества.
        7.2 Система стандартов по ЕИС не исключает действия нормативно-правовых актов, содержащих государственные требования к ИС, в том числе норм и правил, утвержденных органами исполнительной власти в пределах их полномочий. 
        7.3 Система стандартов по ЕИС не должна противоречить и не отменяет действия положений, изложенных в стандартах, нормативно-технических и организационно-распорядительных документах Общества по данному направлению деятельности.
        7.4 Обозначение стандарта ЕИС состоит из индекса (СТО), регистрационного номера, первые две цифры с точкой (50.) определяют принадлежность стандарта к системе стандартов по ЕИС, последующие цифры – порядковый номер стандарта. Стандарты ЕИС должны иметь групповой заголовок: «Единая информационная система ПАО «Сургут-нефтегаз».
        7.5 Для обеспечения требований системы стандартов по ЕИС и исполнения процессов выполняется разработка, актуализация и введение в действие ОТД ЕИС.
            7.5.1 Объектом регламентации ОТД  ЕИС являются: 
- ИС; 
- ИТ; 
- программное средство, программные продукты, ПО; 
- работы и услуги по автоматизации деятельности.
            7.5.2 Настоящим стандартом регулируется применение ОТД  ЕИС в следующих областях:
? регламентация исполнения процессов (технологических, информационных) и услуг, направленных на обеспечение автоматизации деятельности Общества;
? определение правил, норм, требований к оперативному управлению деятельностью в области автоматизации;
- регламентация процессов управления ИТ-решениями;
- обеспечение функционирования ИТ-решений;
- определение и регламентация работы Пользователей (правила, нормы и требования) в ИС; 
- регламентация обслуживания ИС;
? организация документирования ИТ-решений;
? формирование организационно-технических мероприятий, планов работ по внедрению ИТ-решений;
? ведение информационно-справочных данных, связанных с автоматизацией деятельности Общества.
            7.5.3 ОТД ЕИС должны быть отнесены к одному из следующих видов с соответствующим обозначением: 
- порядок исполнения автоматизированного бизнес-процесса (ПБП) – документ, содержащий функции участников, состав и последовательность работ с ИТ-решениями, направленными на получение результата в рамках осуществляемой деятельности;
Примеры
1 Порядок формирования электронного архива первичных документов по продаже нефтепродуктов.
2 Порядок работы в информационной системе «Исполнение бизнес-процессов» при оформлении заявок на обслуживание.
- порядок исполнения ИТ-процесса (ПИТ) – документ, включающий в себя описание состава работ, функций участников, роли, условия и механизмы исполнения ИТ-процессов;
Примеры
1 Порядок ведения каталога ИТ-решений ПАО «Сургутнефтегаз».
2 Порядок управления изменениями информационных систем.
3 Порядок выполнения работ по проектированию ИТ-решений.
- руководство (РУ) – документ, включающий в себя правила, нормы, требования, рекомендации по исполнению определенных действий или последовательности действий, особенностей объекта или технологических условий.
Примеры
1 Руководство по загрузке данных о полномочиях пользователей в систему SAP HANA из системы SAP IdM.
2 Руководство по ведению данных о пользователе в ИС SAP IdM.
3 Руководство по настройке ИС «Исполнение процессов» при создании интеграционных сценариев с другими ИС».
4 Руководство по выполнению разработок в среде ABAP.
5 Руководство по оформлению и согласованию документа «Технические требования».
            7.5.4 Регистрацию ОТД осуществляет управление информационных технологий Общества путем присвоения ОТД регистрационного номера. Регистрационный номер каждого документа имеет вид (таблица 1): 
{Тип}{Номер}.ОТД–{Год}{/Дополнительная кодировка}
Пример – РУ12.ОТД–2019 или ПБП17.ОТД–2019/2
Таблица 1 – Обозначения регистрационного номера 
Обозначение
Пояснения к заполнению
{Тип}
Указывается краткое обозначение вида ОТД, записываемое прописными буквами «ПБП», «ПИТ» или «РУ» (см. 7.5.3)
{Номер}
Указывается сквозной порядковый номер
ОТД
Указывается текст «ОТД» заглавными буквами после порядкового номера, отделенный точкой
{Год}
Указываются отделенные тире от текста «ОТД» четыре цифры года направления окончательной версии на согласование. Начальное значение содержит ожидаемый год введения в действие ОТД
{/Дополнительная кодировка}
Текст, содержащий дополнительную кодировку, указывается в правой части регистрационного номера, отдельный символом «/» от {Год}
            7.5.5 Координацию работ по разработке, изменению, вводу или отмене ОТД осуществляет начальник управления информационных технологий Общества (далее – Координатор ОТД). 
            7.5.6 Разработка ОТД осуществляется управлением, отделом, службой аппарата управления и структурным подразделением Общества, выполняющими работы по управлению, внедрению и применению ИТ в Обществе. 
            7.5.7 Разрабатываемые, пересматриваемые ОТД утверждают и вводят в действие следующими способами:
- утверждается как самостоятельный документ и направляется для исполнения приложением к организационно-распорядительным документам в адрес участников (управлений, отделов, служб аппарата управления и структурных подразделений Общества) с указанием даты ввода в действие;
- вводится в действие указанием Общества, подписанным заместителем генерального директора Общества по направлению деятельности, или приказом Общества (решение принимает Координатор ОТД) с указанием даты ввода в действие;
- вводится в действие приказом структурного подразделения Общества с указанием даты ввода в действие. ОТД, введенный в действие на уровне структурного подразделения Общества, обязателен к применению только для организационных единиц данного структурного подразделения Общества.
            7.5.8 Ведение фонда ОТД осуществляет управление информационных технологий Общества, которое определяет процессы инициации, разработки, согласования, ввода в действие, размещения, актуализации и отмены ОТД. Реализация данных процессов в ИС (в том числе в электронном виде) осуществляется по согласованию с Координатором ОТД.
            7.5.9 Срок действия ОТД устанавливается не более трех лет. При необходимости внесения изменений, дополнений, корректировок в ОТД производится его пересмотр. В Фонде ОТД отражается актуальное состояние. 
            7.5.10 Требования к оформлению ОТД аналогичны требованиям к оформлению нормативно-технических документов (раздел 6 СТО 4).
    8 Группы ролей и участники
        8.1 Для управления ИТ-решениями принята ролевая модель участников процессов согласно рисунку 1. 
        8.2 Группы ролей в бизнес-области:
- группа «Владельцы бизнес-процессов», отвечающая за исполнение и результаты бизнес-процессов и их соответствие намеченным целям;
- группа «Пользователи», использующая ИТ-решения для выполнения своих функций в рамках бизнес-процесса;
- центральная группа «Управление ИТ-решениями», обеспечивающая эффективное взаимодействие с ИТ-областью, отвечающая за общее планирование, определение требований, постановку задач для проектирования, авторский контроль для дальнейшего сопровождения 
ИТ-решений. 
Примечание – Группы «Владельцы бизнес-процессов» и «Пользователи» выступают в качестве Заказчиков ИТ-решений. Группы «Управление ИТ-решениями», «Управление приложениями», «Управление разработками», «Эксплуатация и сопровождение», «Системное администрирование», «Управление ИТ-инфраструктурой» выступают в качестве Исполнителей работ в рамках жизненного цикла ИТ-решений.


Рисунок 1 – Ролевая модель управления ИТ-решениями
        8.3 Группы ролей в ИТ-области, входящие в состав ИТ-служб:
- группа «Управление приложениями», обеспечивающая взаимодействие между группами ролей бизнес-области и ИТ-области в части проектирования и внедрения ИТ-решений, интеграции данных, приложений, компонентов инфраструктуры; 
- группа «Эксплуатация и сопровождение», обеспечивающая функционирование ИТ-решений в заданных условиях в интересах Пользователей и Владельцев бизнес-процессов, инициирующих внесение изменений в ИТ-решения; 
- группа «Системное администрирование», обеспечивающая контроль работоспособности и администрирование компонентов ИС; 
- группа «Управление разработками», обеспечивающая реализацию ПО;
- группа «Управление ИТ-инфраструктурой», обеспечивающая создание и поддержание ИТ-инфраструктуры.
Приложение А
(справочное)

Единая информационная система Общества



















Приложение Б
(справочное)

Процессы управления жизненным циклом ИТ-решения


Начальник управления 
информационных технологий 
ПАО «Сургутнефтегаз»	Р.Д.Гимранов

Начальник отдела мониторинга 
и безопасности информационных 
систем управления 
информационных технологий
ПАО «Сургутнефтегаз» 	А.М.Шевчук

Начальник отдела корпоративной 
архитектуры управления 
информационных технологий 
ПАО «Сургутнефтегаз»	В.А.Агиевич


СОГЛАСОВАНО

Главный инженер – 
первый заместитель генерального 
директора ПАО «Сургутнефтегаз» 	А.Н.Буланов

Первый заместитель генерального
директора ПАО «Сургутнефтегаз»
по финансовым вопросам и налогам   	В.Г.Баранков

Начальник управления криминальной 
и информационной безопасности 
ПАО «Сургутнефтегаз»	А.Е.Григоров

Начальник правового управления
ПАО «Сургутнефтегаз»	Л.А.Логиновская

Начальник управления делами
ПАО «Сургутнефтегаз»	О.И.Романенко

СОГЛАСОВАНО

Должность
Инициалы, фамилия
Подпись
Дата
Заместитель главного инженера – начальник технического управления ПАО «Сургутнефтегаз»
Д.Н.Олейник

17.07.2020
Заместитель начальника технического управления ПАО «Сургутнефтегаз» по новой технике и технологии – начальник технического отдела
М.А.Краснов

10.07.2020
Начальник ПУ «СургутАСУнефть»
С.В.Кононов

10.07.2020
Директор 
Инженерно-экономического внедренческого центра
И.А.Камышин

23.06.2020
Главный инженер Инженерно-экономического внедренческого центра
В.И.Холомеев

23.06.2020
Начальник отдела стандартизации и патентоведения Инженерно-экономического внедренческого центра
Е.А.Базалев

21.06.2020











Распечатано из системы DIRECTUM 18.07.2020 в 09:41
Инженер I категории группы стандартизации
отдела стандартизации и патентоведения ИЭВЦ/Е.С.Бачинская__________
ЛИСТ РЕГИСТРАЦИИ ИЗМЕНЕНИЙ СТО 50.0–2020

Номер измене-ния
Номера листов, страниц
Номер приказа, дата
Срок
введения изменения
Дата
внесения
изменения
Подпись

изменен-ных
заменен-ных
новых
аннули-рованных




1
2
3
4
5
6
7
8
9











'''
```
